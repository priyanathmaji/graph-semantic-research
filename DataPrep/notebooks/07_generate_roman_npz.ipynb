```python
#pip install datasets
from datasets import load_dataset
```

```python
dataset_wiki = load_dataset(
    "wikipedia", 
    "20220301.en", 
    split="train", 
    streaming=True, 
    revision="refs/convert/parquet" # add this to bypass security
)

target_title = "Roman Empire"
roman_empire_data = None

for article in dataset_wiki:
    if article['title'] == target_title:
        roman_empire_data = article
        break

if roman_empire_data:
    print(f"Found: {roman_empire_data['title']}")
    print("-" * 30)
    # This prints the first 1000 characters of the raw text
    print(roman_empire_data['text'][:1000] + "...")
else:
    print("Article not found in this dump.")
```
