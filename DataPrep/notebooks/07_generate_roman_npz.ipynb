```python
#pip install datasets
from datasets import load_dataset
```

```python
# Roman Empire was not available in the wikipedia, so had to download from a mirror
dataset_wiki = load_dataset("olm/olm-wikipedia-20221220")
```

```python
for article in dataset_wiki['train']:
    if article['title'] == "Roman Empire":
        print(f"Title: {article['title']}")
        print(f"Text 500: {article['text'][:500]}")
        break
```

```python
for article in dataset_wiki['train']:
    if article['title'] == "Roman Empire":
        with open("roman_empire_wikipedia_2022.txt", "w", encoding="utf-8") as f:
            f.write(article['text'])
        
        print("Success! The full text has been saved to 'roman_empire_wikipedia_2022.txt'")
        break
```

```python
with open("../data/roman_empire_wikipedia_2022.txt", "r", encoding="utf-8") as f:
    roman_empire = f.read()
```

```python
import spacy
import networkx as nx

#python -m spacy download en_core_web_md
nlp = spacy.load("en_core_web_md")
doc = nlp(roman_empire)

G = nx.DiGraph()

for i, token in enumerate(doc):
    # Sequential edges (chain graph)
    if i < len(doc) - 1:
        G.add_edge(i, i + 1)
    
    # Syntactic dependency edges
    if token.head.i != token.i:
        G.add_edge(token.i, token.head.i)

print(f"Graph built with {G.number_of_nodes()} nodes and {G.number_of_edges()} edges.")

```
